# Importing Datasets

```python
df_train=pd.read_csv('/kaggle/input/random-linear-regression/train.csv')
df_test=pd.read_csv('/kaggle/input/random-linear-regression/test.csv')
```

# Data Preprocessing

```python
df_train.head()
df_train.info()
```

```python
# from the above code we see that y has 1 null value so we will either drop the row or replace the 
# value with mean of colum 'y'
# Here we will drop the row since we have large dataset 

df_train.dropna(inplace=True)
df_train.info()
```

```python
df_test.info()
```

# Data Visualization

Lets find the best model by plotting graph between 'x' and 'y'

```python
plt.scatter(df_train['x'],df_train['y'],color='g',marker='.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

**We find that there is linear relationship between 'x' and 'y' and that linear regression is most suitable for the data**

# Building and Training the Model

```python
X_train=df_train[['x']]
y_train=df_train[['y']]
X_test=df_test[['x']]
y_test=df_test[['y']]
```

<h3><b>Linear Regression</b></h3>

```python
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
regr.fit(X_train,y_train)
```

We will find the accuracy of our model using r2_score

```python
from sklearn.metrics import r2_score
yhat=regr.predict(X_test)
print("The score of model is :",r2_score(yhat,y_test))
```

***The accuracy of our model comes out to be 98.83 so we see Linear Regression is the best model for the dataset***

**Let's Plot graph between test points and our prediction**

```python
plt.figure(figsize=(10,8))
plt.scatter(X_test,y_test,color='g',marker='.',label='Test Points')
plt.plot(X_test,yhat,color='blue',label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower right')
plt.title('X vs Y')
plt.show()
```
