# Creating Functions

Functions are pretty handy for turning repetitive tasks into a single line of code, let us now create our own function:

```python
def say_hello():
    '''
    Prints hello to the terminal
    '''
    print('Hello!')
```

Creating a function and executing a function are different things. We create the `say_hello()` function, but we did not execute it.

Let's execute it now!

```python
say_hello()
```

Functions can also take arguments, they 

```python
def double_value(x):
    '''
    Multiplies received value by two
    '''
    result = x * 2
    return(result)
```

Functions are great for encapsulation multiple general tasks, and can also be reused to increase productivity
We can join functions with data - this is known as a **class**!

```python
class Doggo():
    breed          = 'husky'
    favorite_place = 'beach'
    
    def woof(self):
        print('WoOF!')
```

Take the doggo class for example, it has **attributes** which is information of it's breed and favorite place to go, as well as **methods** which are functions that it can execute, such as woofing in this case.

Let's now create an instance of that class and check our doggo

```python
freddie = Doggo()
freddie.breed
```

```python
freddie.woof()
```

# Creating Loops

We can also create loops to repeat a certain action multiple times, here's how it works:

```python
for i in range(10):
    print(i)
```
