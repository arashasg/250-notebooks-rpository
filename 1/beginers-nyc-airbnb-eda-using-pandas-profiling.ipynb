# **Pandas Profiling**

It is an open source library used for EDA,which can be done in few lines of code.

```python
#to get a simple report with pandas profiling
from pandas_profiling import ProfileReport
report = ProfileReport(train, title='Pandas Profiling Report')
```

Reports generated using pandas profiling provides detailed report of the dataset. It shows the data distribution correlation,missing values etc with just a single line of code.

There are 2 methods to generate a report:
1. to_widgets() -  this generates the report in the notebook
2. to_file("output.html") - this generates a report in a html file and saved in the same location where the notebook is saved. we can send this to the clients for their easy understanding.

```python
#to create a report use to_widgets
report.to_widgets()
```

```python
report.to_file('output.html')
```
