# **Now we will be processing our data and forming a model for it**

```python
train_data.dropna(axis=0, subset=['Survived'], inplace=True)
y= train_data.Survived
train_data.drop(['Survived'], axis=1, inplace=True)
```

```python
train_data
```

```python
train_data.isnull().nunique()
```

Joining both the train and test data sets

```python
df = pd.concat([train_data,test_data])
```

```python
df
```

```python
df.isnull().sum()
```

Refining our NAN values with simple imputer

```python
from sklearn.impute import SimpleImputer 
#now we will be imputing our data
imputer = SimpleImputer(strategy='most_frequent')
```

```python
df_imputed = pd.DataFrame(imputer.fit_transform(df))
df_imputed.columns = df.columns
```

```python
df = pd.get_dummies(df_imputed)
```

```python
df
```

```python
train = df.iloc[:891,]
test =df.iloc[891:,]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_valid,y_train,y_valid =train_test_split(train,y,train_size=.80,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr =LinearRegression()
lr.fit(X_train,y_train)
accuracy_lr = lr.score(X_valid,y_valid)
```

```python
accuracy_lr
```

```python
from sklearn.ensemble import RandomForestClassifier
model =  RandomForestClassifier(n_estimators=100, random_state=1)
model.fit(X_train, y_train)
acc = model.score(X_valid,y_valid)
```

```python
acc
```

```python
from sklearn.svm import SVC
models=SVC()
models.fit(X_train,y_train)
acu= models.score(X_valid,y_valid)
acu
```

Therefore we will use RandomForetClassifier

Predicting our model on test data

```python
pred = model.predict(test)
```
